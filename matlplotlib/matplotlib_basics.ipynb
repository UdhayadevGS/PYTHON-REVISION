```python
import matplotlib.pyplot as plt
```

```python
# plot

# Example rows values - years
x = [2000,2001,2002,2003,2004,2006] #notice 2005 is missing
#example column values - student's marks over years
y=[10,20,30,20,10,30]
#markers mark wherever the data is present in the graph
plt.plot(x,y,color='blue',linestyle='--',marker='o',label='John')

 #LET THIS BE MARKS OF ANOTHER STUDENT
z=[10,25,40,65,85,95]
plt.plot(x,z,color='red',marker='x',label='Greg')
plt.xlabel('Years')
plt.ylabel('Student Marks')

plt.title('Students Marks vs Years')

plt.legend() #you dont need to mention labels inside as they are already defined in those two functions

plt.show()
```

```python
#scatter

x = [1,2,3,4,5,6,7,8]
y = [5,2,4,2,1,4,5,2]

plt.scatter(x,y,label='skitscat',color='k',marker='*',s=50,alpha=0.9) #label acts as the legend , colour gives the colour - check colour code,
# s controls size of the markers and alpha controls the transpareny of the markers (0= invisible , 1 is opaque)
plt.xlabel('x')
plt.ylabel('y')
plt.title('Scatter Plot')
plt.legend()
plt.show()
```

```python
#vertical bar graphs
x=['A','B','C','D'] #STUDENT NAMES
y=[20,40,10,30] #STUDENT WEIGHTS
colors = ['red','green','yellow','blue'] #COLOUR OF THE BAR GRAPHS
import pandas as pd #WE USE PANDAS FOR CREATING A DATAFRAME AND SORTING IT BY WEIGHT
df = pd.DataFrame({'agecolumn':x,'weightcolumn':y,'colorscolumn':colors}) #CREATES A PANDAS DATAFRAME BY INPUTTING DICTIONARY WHICH HAS COLUMN NAMES  AS KEY AND LIST OF DATA AS VALUES
df = df.sort_values(by='weightcolumn') #SORT THE DATAFRAME BY WEIGHT
plt.bar(df['agecolumn'],df['weightcolumn'],color=df['colorscolumn']) #PLOTS THE BAR GRAPH



plt.xlabel('STUDENT NAME')
plt.ylabel('Weight')
plt.title('STUDENT NAME VS WEIGHT')
plt.grid()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
emp = pd.read_csv('/content/employee_sales_2024_2025.csv')

fig,(ax1,ax2) = plt.subplots(1,2,figsize=(20,7)) #FIG IS THE ENTIRE CANVAS , AXES ARE GRAPHS WITHIN THE CANVAS , SUBPLOTS(1,2) CREATES SUBPLOTS WITH 1 ROW AND 2 COLUMNS

#emp2024 is a subset of emp with name,sales_2024 excluding nan values
emp2024 = emp.dropna(subset='sales_2024').copy()
#sorting values based on sales
emp2024 = emp2024.sort_values(by='sales_2024')
#plot the bar graph and store the bars in a variable for later use - labelling
bars1=ax1.bar(emp2024['employee'],emp2024['sales_2024'],color=emp2024['color'])
#now we create the labels for each bar
#we cannot iterate through many columns at once so we first zip columns together , then we unpack them , and then create labels
#we use f strings to create the labels
labels1=[f"{name}:{sale}" for name,sale in zip(emp2024['employee'],emp2024['sales_2024'])]
#now we can label the bars
ax1.bar_label(bars1,labels=labels1)
ax1.set_xlabel('Employee names')
ax1.set_ylabel('Number Of Sales')
ax1.grid()
ax1.set_title('2024 SALES BY EMPLOYEES')

#now doing the same for axis two - with data of 2025

emp2025 = emp.dropna(subset='sales_2025').copy()
emp2025 = emp2025.sort_values(by='sales_2025')
bars2=ax2.bar(emp2025['employee'],emp2025['sales_2025'],color=emp2025['color'])
labels2=[f"{name}:{sale}" for name,sale in zip(emp2025['employee'],emp2025['sales_2025'])]
ax2.bar_label(bars2,labels=labels2)
ax2.set_xlabel('Employee names')
ax2.set_ylabel('Number Of Sales')
ax2.grid()
ax2.set_title('2025 SALES BY EMPLOYEES')

fig.show()
```
