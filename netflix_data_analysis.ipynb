```python
import pandas as pd
```

```python
test = pd.read_csv('/content/netflix_titles.csv')
```

```python
test.head()
```

```python
# NUMBER OF MOVIES AND TV SERIES RELEASED EVERY YEAR

typeyear = test.groupby(['type','release_year']).size()
#groupby('type') -> groups the dataset by the type (Movie/Series) and release year
#.size() -> gives the count of movies released in that particular year for Movie/Series
typeyear
```

```python
#now i want to check the average duration of movies for each year

#first i will get only groups that are movies
movies = test[test['type'] == 'Movie'].copy()
#now i will drop values with have NaN
movies =  movies.dropna(subset=['duration']).copy()
#now i want to replace the min in movies mins to do calcualtions
movies['duration_minutes'] = movies['duration'].str.replace(' min', '').astype(int)
#now i will group movies according to year and find the mean of the average minutes
movies_mean = movies.groupby('release_year')['duration_minutes'].mean()
display(movies_mean)


movies_mean_median = movies.groupby('release_year')['duration_minutes'].agg(['mean','median'])
movies_mean_median.tail()
```

```python
#now more advanced i want to extract number of seasons and duration of movies for each year
#first i will drop all nan values from duration and type
moviesseries =  test.dropna(subset=['duration','type']).copy()
# Create the new columns filled with None (or NaN)
moviesseries['duration_minutes'] = None
moviesseries['no_of_seasons'] = None
# For movies: extract number of minutes from 'duration'
moviesseries.loc[moviesseries['type'] == 'Movie', 'duration_minutes'] = moviesseries.loc[moviesseries['type'] == 'Movie', 'duration'].str.replace(' min', '').astype(int)
# For TV shows: extract number of seasons (handle singular and plural)
moviesseries.loc[moviesseries['type'] == 'TV Show', 'no_of_seasons'] = moviesseries.loc[moviesseries['type'] == 'TV Show', 'duration'].str.replace(' Seasons', '').str.replace(' Season', '').astype(int)
moviesdurmean=moviesseries.groupby('release_year')['duration_minutes'].mean()
display(moviesdurmean)
seasonsmean=moviesseries.groupby('release_year')['no_of_seasons'].mean()
display(seasonsmean.tail(20))
```

```python
#NOW FOR TOP COUNTRIES PRODUCING CONTENT - AS A WHOLE
countrycount=test.dropna(subset=['country']).copy()
countrycount=countrycount.groupby('country')['country'].count().sort_values(ascending=False)
```

```python
year_wise_rating = test.dropna(subset=['rating']).copy()
year_wise_rating = year_wise_rating.groupby(['release_year','rating']).size()
year_wise_rating
```

```python
country_wise_rating = test.dropna(subset=['country','rating']).copy()
country_wise_rating = country_wise_rating.groupby(['country','rating']).size()
country_wise_rating.loc['India']

```

```python
#NOW, OBTAINING THE NUMBER OF RELEASES FOR EACH GENRE FOR EACH YEAR
genre_by_year = test.dropna(subset=['release_year','listed_in']).copy() #drops rows where release year or listed_in is na , stores the filtered table to genre_by_year
genre_by_year['listed_in'] = genre_by_year['listed_in'].str.split(', ') #splitting the elements in listed_in to a list
genre_by_year_exploded=genre_by_year.explode('listed_in')  #explodes the list in listed_in so that each element in that list will have a column of its own
genre_by_year=genre_by_year_exploded.groupby(['release_year','listed_in']).size()

genre_by_year.head(10)
```

```python
#NOW, TO GET THE DIRECTORS WITH MOST NUMBERS OF RELEASES
dir = test.dropna(subset=['director']).copy()
dir = dir.groupby('director').size().sort_values(ascending=False)
dir.head(10)
```

```python
#FREQUENTLY APPEARNING CAST
freq_cast = test.dropna(subset=['cast']).copy()
freq_cast = freq_cast['cast'].str.split(', ') #to convert cast strings to a list
freq_cast_explode = freq_cast.explode() #converts the list elements to seperate rows and duplicates values automatically to each of them
freq_cast = freq_cast_explode.groupby(freq_cast_explode).size().sort_values(ascending=False)
freq_cast.head(10)
```

```python
country_wise_production = test.dropna(subset=['country','release_year']).copy()
country_wise_production = country_wise_production.groupby(['country','release_year']).size().sort_values(ascending=False)
country_wise_production.head(10)
```
