```python
import pandas as pd
```

```python
coffee = pd.read_csv('coffee.csv') #reads file coffee

```

```python
coffee.head(5)  #gives first 5 row data values of coffee, doesnt save to coffee unless you mention coffee = coffee/head()
```

```python
coffee.tail(5)  #gives last 5 row data values of coffee
```

```python
coffee.loc[0:5] #loc format is name.loc[rows,columns] ;  this gives rows with - index from 0 to 5
```

```python
coffee.iloc[0:5,[0,2]] #gives rows with - index from 0 to 4 (5 is exclusive) , and gives columns indexed 0 and 2 - which are Day and Units Sold
```

```python
coffee.loc[0:4] #gives all rows data from index 0 to 4
```

```python
coffee.loc[0:4,["Day","Units Sold"]] # gives all rows data from 0 to 4 and only columns Day and Units Sold
```

```python
coffee = coffee.set_index('Day') #sets Day column as the index
```

```python
coffee # we can see Day is now the index for rows

```

```python
coffee.loc['Monday':'Friday'] #now that days column is the index, we can access data with days
```

```python
coffee.reset_index() #resets index back to default
```

```python
#setting values
coffee.loc[0:2,"Units Sold"]=[100,200,300] #CHANGING VALUES
coffee.loc[0,'Day']='IdontcareDay'
coffee.loc[[3,4],['Coffee Type','Units Sold']] =[['NewEspresso',99],['NewLatte',100]]
coffee
```

```python
coffee.sort_values(['Units Sold','Day'],ascending=[False,True]) #sorting values - DESCENDING FOR UNITS SOLD , ASCENDING FOR DAY
```
